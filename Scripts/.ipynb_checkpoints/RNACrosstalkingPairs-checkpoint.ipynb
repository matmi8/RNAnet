```python
import numpy as np
import matplotlib.pyplot as mpl
import os, sys



import scipy.optimize 
from scipy.ndimage import measurements

sys.path.append("../Functions/")
import Network_functions as NF
import Susceptibility_functions as SF

%matplotlib tk
```

```python
def orderMatrix(a, x, y):

    m,n = np.shape(a)
    
    tmp = np.column_stack((a[x,y], x, y))
    tmp2 = tmp[np.argsort(tmp[:,0])]
    
    res = tmp2[:,1] + 1j*tmp2[:,2]
                          
    return(res, tmp2[:,0])
```

# Importing data

```python
#Tr = [0.1,0.3,0.4,0.6,0.8,0.9, 1.,1.1, 1.2,1.5,2.]
Tr = [0.1,0.6,0.9,1.2,1.5,2.]
Bd = 2

Ntr = len(Tr)


### Number of couples to extract  with highest mean(X) and K_ij = 0...
pos = 15


### creating a fiticiuos X in order to know the K_ij.. XD
links = NF.LinksFunc(filename="../Networks/CLASH_whole_links.dat", Bd=Bd, m0=6.75)
M,N = NF.SpeciesNumbers(links=links)
X,K = SF.X_K_ij_func(links=links, lam=0.01, m=np.ones(M), m_=np.ones(M), n=np.ones(N), n_= np.ones(N))
np.fill_diagonal(K, -1) 

### Finding all i,j where K_ij > 0
posX_0, posY_0 = np.where(K == 0)

### Finding all i,j where K_ij >= 0 (excluding diagonal elements only)
posX_all, posY_all = np.where(K >= 0)

### Finding all i,j where K_ij >= 0 (only direct contact)
posX_k1, posY_k1 = np.where(K > 0)


### freeing memory...
X = 0
K = 0

### Finding number of X_ij couples with K_ij = 0
L_0 = len(posX_0)

### Finding number of X_ij couples (offdiagonal)
L_all = len(posX_all)

### Finding number of X_ij couples with K_ij >0 
L_k1 = len(posX_k1)
```

# K=0 CASE

```python
### allocating matrices that countain couple ij indices and Xs
RNA_vet_ij_0 = np.zeros((L_0,Ntr), dtype="complex")
RNA_vet_X_0 = np.zeros((L_0,Ntr), dtype="float32")


maxX = np.zeros(Ntr)

### Loading data...
for i in range(Ntr):
    filename = "/home/mattia/RNAnet/ResultsPaper/RNA_PairsX/meanX_CLASH_whole_Tr_%.3e_Bd_%1.0d.dat"%(Tr[i], Bd)
    print(filename)
    X = np.loadtxt(filename)
    
    ## Finding the maximum X among all X (excluding off-diagonal terms)...
    maxX[i]  = np.max(X[posX_all, posY_all])
    
    tmp, Xtmp = orderMatrix(X, posX_0, posY_0)
    RNA_vet_ij_0[:,i] = tmp
    RNA_vet_X_0[:,i] = Xtmp
    
    
### Freeing memory...
X = []
tmp = []
Xtmp = []
```

```python
## Checking the ordering of the couples having K_ij = 0! XD
print("Top %d Xs"%pos)
print(RNA_vet_X_0[-pos:,])

print("Max Xs")
print(maxX)


### Computing number of couples:
print("Total number of couples is", M*(M-1))
print("Couples at distance > 1 are" , len(posX_0))
print("Couples at distance 1 are", M*(M-1) - len(posX_0))


### Assembly top 15 X couples susceptibilities and the maximum X in a matrix
ResX = np.row_stack([maxX, RNA_vet_X_0[-pos:,]])
np.savetxt("/home/mattia/RNAcouples_X.dat",ResX)
```

# Extracting top 15 X couples (K=0)

```python
### Printing common couples 
for i in range(Ntr-1):
    print("CV %.1f --> CV %.1f"%(Tr[i], Tr[i+1]))
    print(np.intersect1d(RNA_vet_ij_0[-pos:,i], RNA_vet_ij_0[-pos:,i+1]))
    print("  ")

print("####################")



Tr_vet = np.array(Tr)
print(Tr_vet)

I = []
J = []
CV = []

for a in range(Ntr):    
    I = np.concatenate([I,np.real(RNA_vet_ij_0[-pos:,a])])
    J = np.concatenate([J,np.imag(RNA_vet_ij_0[-pos:,a])])
    CV = np.concatenate([CV,np.repeat(Tr[a], pos)])
    
Couples = np.column_stack([I,J,CV])

np.savetxt("/home/mattia/RNAcouples_.txt",Couples)

print("Couples")
print(Couples)


```

```python
### NB For us, percentile 0 corresponds to the highest X couples!!

Ndiv = 6

Ncouples, tmp = np.shape(RNA_vet_ij_0)
M = int(Ncouples/Ndiv)

print("Ncoupples = ",Ncouples, "   M = ",M)

ResPerc = np.zeros((Ndiv*(Ntr-1),2+Ndiv))

BINS = np.linspace(-30,1, 100)


## cyclying over Tr from 0 to Ntr-1
for i in range(Ntr-1):
    print("CV = ", Tr[i], "-->", Tr[i+1])
    
    ## cyclying over divisions of Tr[i]...
    for a in range(Ndiv):
        #mpl.figure()
   
        ResPerc[i*Ndiv + a, 0] = a
        ResPerc[i*Ndiv + a, -1] = Tr[i] 
        
        col_i = RNA_vet_ij_0[-(M*(a+1)):-(M*a+1),i]
        col_i_x = RNA_vet_X_0[-(M*(a+1)):-(M*a+1),i]
        
        if(a == (Ndiv-1)):
            col_i = RNA_vet_ij_0[:-(M*a+1),i]
            col_i_x = RNA_vet_X_0[:-(M*a+1),i]
                
        #mpl.hist(np.log10(col_i_x), bins=BINS, label = "i")
        ## cyclying over divisions of column Tr[i+1]...
        for b in range(Ndiv):
            
            col_j = RNA_vet_ij_0[-(M*(b+1)):-(M*b+1),i+1]
            col_j_x = RNA_vet_X_0[-(M*(b+1)):-(M*b+1),i+1]
            
            if(b == (Ndiv-1)):
                col_j = RNA_vet_ij_0[:-(M*b+1),i+1]
                col_j_x = RNA_vet_X_0[:-(M*b+1),i+1]
                
            inter = np.intersect1d(col_i,col_j)
            
            perc = (len(inter)/float(M+1))
            ResPerc[i*Ndiv+a, 1+b] = perc
            print("percentile  ", a, "in",b, "==>", perc )
            #mpl.hist(np.log10(col_j_x), bins=BINS, label = "j=%d"%b)
        #mpl.legend()
        #mpl.show()
        
#np.savetxt("/home/mattia/PercentileCouple_farX_%d_divisions.dat"%Ndiv, ResPerc, fmt="%.3e")
```
